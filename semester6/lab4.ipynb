Краткое описание

 * team_n - название команды, играющей за роль n = 1, 2.

 * tn_points - Количество набранных очков.

 * h2h win perc: доля побед в прошлых играх против другой команды. Равно 0,5, если предыдущих противостояний не было.

tn_playerk_ - Игроки упорядочиваются от 1 до 5 в соответствии с их рейтингом k.

 * rating: метрика качества игрока.

 * impact: общее влияние игрока на матчи.

 * kdr: соотношение общего числа убийств и общего числа смертей.

 * mr: средний урон за раунд.

 * kpr: среднее число убийств за раунд.
 * apr: среднее число помощей в убийстве за раунд.
 * dpr: среднее число смертей за раунд.
 * spr: среднее число спасённых сокомандников за раунд.
 * opk ratio: соотношение общего числа первых убийств к общему числу первых сметрей за раунды.
 * opk rating: hltv показатель первых убийств игрока.
 * wins perc after fk: доля побед после первого убийства этого игрока.
 * fk perc in wins: доля первых убийств игрока в выйгранных раундах.
 * multikill perc: доля раундов, в которых игрок совершил >= 2 убийств.
 * rating at least one perc: доля матчей, в которых у игрока рейтинг как минимум 1.0.
 * is sniper: является ли первое или второе оружие игрока по количеству убийств "awp".
 * clutch win perc: доля побед в схватках с другим игроком один на один.

# Реализация и обучение базовой RBF-сети

## Реализация

```python
import torch
import torch.nn as nn
from sklearn.cluster import KMeans



class RBF(nn.Module):
    def __init__(self, in_dim, n_centers, out_dim, rbf_type='gs'):
        assert rbf_type in ['gs', 'mq', 'inv_mq', 'mtps']

        super(RBF, self).__init__()
        self.n_centers = n_centers
        self.centers = nn.Parameter(torch.randn(n_centers, in_dim))
        self.sigmas = nn.Parameter(torch.ones(n_centers))
        self.rbf = (self.gaussian if rbf_type == 'gs' else self.multiquadric if rbf_type == 'mq' 
                    else self.inverse_multiquadric if rbf_type == 'inv_mq' else self.mod_thinplate_spline)
        self.linear = nn.Linear(n_centers, out_dim)
    
    def _kmeans_init(self, x):
        kmeans = KMeans(n_clusters=self.n_centers)
        kmeans.fit(x.cpu().numpy() if isinstance(x, torch.Tensor) else x)
        with torch.no_grad():
            self.centers.data = torch.tensor(kmeans.cluster_centers_, dtype=torch.float32, device=x.device)
    
    def _sigmas_init(self, init_type='global', m=1):
        assert init_type in ['global', 'nn']

        with torch.no_grad():
            if init_type == 'global':
                self.sigmas.data = nn.Parameter(torch.max(torch.cdist(self.centers, self.centers)) 
                                        / (2*self.n_centers)**0.5)
            else:
                self.sigmas.data = nn.Parameter(
                    torch.mean(
                        torch.topk(
                            torch.cdist(self.centers, self.centers).fill_diagonal_(float('inf')), 
                            m, largest=False).values, 
                    dim=1)
                )
        
    def gaussian(self, x):
        return torch.exp(-torch.div(
            torch.pow(torch.cdist(x, self.centers), 2), 
            2*torch.pow(self.sigmas, 2)))
    
    def multiquadric(self, x):
        return torch.sqrt(
            torch.div(torch.pow(torch.cdist(x, self.centers), 2), 
                      2*torch.pow(self.sigmas, 2)) 
            + torch.ones_like(self.sigmas))
    
    def inverse_multiquadric(self, x):
        return 1.0 / self.multiquadric(x)
    
    def mod_thinplate_spline(self, x):
        return (torch.pow(torch.div(torch.cdist(x, self.centers), self.sigmas), 2) 
                * torch.log(torch.div(torch.cdist(x, self.centers), self.sigmas)))
    
    def forward(self, X):
        rbf_out = self.rbf(X)
        return self.linear(rbf_out)

    def fit_lr(self, x, y):
        with torch.no_grad():
            rbf_features = self.rbf(x)
            rbf_features = torch.column_stack([rbf_features, torch.ones(len(x), device=rbf_features.device)])
            
            pseudo_inverse = torch.linalg.pinv(rbf_features)
            W = pseudo_inverse @ y
            
            self.linear.weight.data = W[:-1].T.clone().detach().requires_grad_(True)
            self.linear.bias.data = W[-1].T.clone().detach().requires_grad_(True)

    def fit_gd(self, x, y, epochs=100, lr=0.01, batch_size=None):
        criterion = nn.MSELoss()
        optimizer = torch.optim.SGD(self.parameters(), lr=lr)
        
        losses = []
        for _ in range(epochs):
            if batch_size is None:
                optimizer.zero_grad()
                outputs = self(x)
                loss = criterion(outputs, y)
                loss.backward()
                optimizer.step()
            else:
                permutation = torch.randperm(x.size(0))
                for i in range(0, x.size(0), batch_size):
                    indices = permutation[i:i + batch_size]
                    X_batch, y_batch = x[indices], y[indices]
                    optimizer.zero_grad()
                    outputs = self(X_batch)
                    loss = criterion(outputs, y_batch)
                    loss.backward()
                    optimizer.step()
            
            losses.append(loss.item())
        
        return losses
```

## Обучение

```python
# Выбор типа RDF слоя
import pandas as pd
import time
from sklearn.metrics import mean_absolute_error, root_mean_squared_error

results = pd.DataFrame(columns=['rbf', 'time', 'rmse', 'mae'])
losses = []
x_train = x_train.cuda()
y_train = y_train.cuda()
x_val = x_val.cpu()
y_val = y_val.cpu()

for rbf_type in ['gs', 'mq', 'inv_mq', 'mtps']:
    rbf = RBF(x_train.shape[1], 16, y_train.shape[1], rbf_type).to('cuda')
    rbf._kmeans_init(x_train)
    rbf._sigmas_init('nn', 4)

    start = time.time()
    losses.append(rbf.fit_gd(x_train, y_train, 100, batch_size=24))
    elapsed = time.time() - start
    
    y_pred = rbf(x_val.cuda()).cpu().detach().numpy()
    results = pd.concat([results, pd.DataFrame({
        'rbf': [rbf_type],
        'time': [elapsed],
        'rmse': [root_mean_squared_error(y_pred, y_val)], 
        'mae': [mean_absolute_error(y_pred, y_val)]
    })], ignore_index=True)
    print(rbf_type, elapsed)
```

```python
# Подбор гиперпараметров обучения
import pandas as pd
import time
from sklearn.metrics import mean_absolute_error, root_mean_squared_error


results = pd.DataFrame(columns=['centers', 'sigmas', 'fit', 'time', 'rmse', 'mae'])
losses = []
x_train = x_train.cuda()
y_train = y_train.cuda()
x_val = x_val.cpu()
y_val = y_val.cpu()

for trn in ['pinv matrix', 'batch', 'mini-batch', 'stochastic']:
    for ctr in ['random', 'kmeans']:
        for sgm in ['global', 'nn']:
            rbf = RBF(x_train.shape[1], 16, y_train.shape[1]).to('cuda')
            if ctr == 'kmeans':
                rbf._kmeans_init(x_train)
            
            rbf._sigmas_init(sgm, 4)

            start = time.time()
            if trn == 'pinv matrix':
                rbf.fit_lr(x_train, y_train)
            elif trn == 'batch':
                losses.append(rbf.fit_gd(x_train, y_train, 100, batch_size=None))
            elif trn == 'mini-batch':
                losses.append(rbf.fit_gd(x_train, y_train, 100, batch_size=24))
            elif trn == 'stochastic':
                losses.append(rbf.fit_gd(x_train, y_train, 100, batch_size=1))
            
            elapsed = time.time() - start
            y_pred = rbf(x_val.cuda()).cpu().detach().numpy()
            results = pd.concat([results, pd.DataFrame({
                'centers': [ctr],
                'sigmas': [sgm],
                'fit': [trn],
                'time': [elapsed],
                'rmse': [root_mean_squared_error(y_pred, y_val)], 
                'mae': [mean_absolute_error(y_pred, y_val)]
            })], ignore_index=True)
            print(ctr, sgm, trn, elapsed)
            
```

# Оптимизация RBF-сети

## Гибридная модель RBF + MLP

```python
import torch.nn as nn


class RBFMLP(RBF):
    def __init__(self, in_dim, n_centers, out_dim, hidden_dims=[64, 32], activation='relu', rbf_type='gs'):
        super().__init__(in_dim, n_centers, out_dim, rbf_type)
        
        layers = []
        prev_dim = n_centers
        for dim in hidden_dims:
            layers.append(nn.Linear(prev_dim, dim))
            layers.append(self._get_activation(activation))
            prev_dim = dim
        
        layers.append(nn.Linear(prev_dim, out_dim))    
        self.linear = nn.Sequential(*layers)
    
    def _get_activation(self, name):
        if name == 'relu':
            return nn.ReLU()
        elif name == 'sigmoid':
            return nn.Sigmoid()
        elif name == 'tanh':
            return nn.Tanh()
        else:
            raise ValueError(f"Unknown activation: {name}")    
```

```python
configs = [
    {'n_centers': 10, 'hidden_dims': [48], 'activation': 'relu'},
    {'n_centers': 30, 'hidden_dims': [96, 48], 'activation': 'tanh'},
    {'n_centers': 50, 'hidden_dims': [32, 20, 8], 'activation': 'sigmoid'},
    {'n_centers': 10, 'hidden_dims': [96, 48], 'activation': 'relu'},
    {'n_centers': 30, 'hidden_dims': [48], 'activation': 'tanh'},
    {'n_centers': 50, 'hidden_dims': [32, 20, 8], 'activation': 'sigmoid'},
    {'n_centers': 10, 'hidden_dims': [32, 20, 8], 'activation': 'relu'},
    {'n_centers': 30, 'hidden_dims': [96, 48], 'activation': 'tanh'},
    {'n_centers': 50, 'hidden_dims': [48], 'activation': 'sigmoid'},
]

results = []
for config in configs:
    model = RBFMLP(
        n_centers=config['n_centers'],
        in_dim=16,
        out_dim=y_train.shape[1],
        hidden_dims=config['hidden_dims'],
        activation=config['activation']
    ).to('cuda')
    model._kmeans_init(x_train[:, :16])
    model._sigmas_init('nn', 4)
    
    model.fit_gd(x_train[:, :16], y_train, epochs=100, batch_size=16)
    y_pred = model(x_val[:, :16].cuda()).cpu().detach().numpy()
    
    results.append({
        'config': config,
        'rmse': [root_mean_squared_error(y_pred, y_val)], 
        'mae': [mean_absolute_error(y_pred, y_val)]
    })
    print(config)
```

## Ансамбль RBF сетей

```python
import numpy as np


class BaggingRBF:
    def __init__(self, in_dim, n_centers, out_dim, n_models=5, device='cpu'):
        self.n_models = n_models
        self.models = [RBF(in_dim, n_centers, out_dim).to(device) for _ in range(n_models)]
    
    def _init_rbf(self, x, sigmas_type='global', m=1):
        for model in self.models:
            model._kmeans_init(x)
            model._sigmas_init(sigmas_type, m)

    def fit(self, x, y, epochs=50, lr=0.01, batch_size=None):
        losses = []
        for i, model in enumerate(self.models):
            # Извлекаем подвыборку
            indices = np.random.choice(x.shape[0], size=x.shape[0], replace=True)
            x_subset = x[indices]
            y_subset = y[indices]
            
            print(f"Training model {i+1}/{self.n_models}")
            model_losses = model.fit_gd(x_subset, y_subset, epochs=epochs, lr=lr, batch_size=batch_size)
            if not losses:
                losses = model_losses
            else:
                losses = [losses[i] + model_losses[i] for i in range(len(losses))]
        
        return [losses[i] / self.n_models for i in range(len(losses))]
    
    def predict(self, x):
        predictions = []
        for model in self.models:
            with torch.no_grad():
                pred = model(x)
                predictions.append(pred)

        return torch.mean(torch.cat(predictions, dim=1), dim=1).unsqueeze(1)
```

```python
rbf = RBF(16, 10, y_train.shape[1]).to('cuda')
rbf._kmeans_init(x_train[:, :16])
rbf._sigmas_init('nn', 4)
rbf.fit_gd(x_train[:, :16], y_train, batch_size=16)

rbf_bagging = BaggingRBF(16, 10, y_train.shape[1], n_models=8, device='cuda')
rbf_bagging._init_rbf(x_train[:, :16], 'nn', 4)
rbf_bagging.fit(x_train[:, :16], y_train, epochs=100, batch_size=16)
```

```python
single_pred = rbf(x_test[:, :16].cuda()).cpu().detach().numpy()
bagging_pred = rbf_bagging.predict(x_test[:, :16].cuda()).cpu().detach().numpy()

print(f'Single RBF: rmse={root_mean_squared_error(y_test, single_pred)}, mae={mean_absolute_error(y_test, single_pred)}')
print(f'RBF Baggiing: rmse={root_mean_squared_error(y_test, bagging_pred)}, mae={mean_absolute_error(y_test, bagging_pred)}')
```

Точность немного увеличилась, но затраты по памяти и времени увеличелись значительно больше, потому нужно либо подбирать гиперпараметры, либо не использовать бэггинг.

```python
def compute_feature_importance(rbf_net):
    weights = torch.abs(rbf_net.linear.weight).mean(dim=0)
    centers = rbf_net.centers 
    std_centers = torch.std(centers, dim=0)  
    
    weighted_importance = weights.unsqueeze(1) * std_centers  
    feature_importance = torch.sum(weighted_importance, dim=0) 
    
    feature_importance /= torch.sum(feature_importance)   
    return feature_importance.cpu().detach().numpy()
```
